```python
# Load data
import pandas as pd
import numpy as np
from datetime import datetime

# Read CSV into a DataFrame
df = pd.read_csv('/Users/berni/Desktop/DESN4003/DESN4003-Respository/TRIALFORCURSOR.csv')
```

```python
# New column to assign random date between 2000 - 2024

# Define date range explicitly
start_date = datetime(2000, 1, 1)
end_date = datetime(2024, 12, 31)

# Convert to Unix timestamps
start_ts = int(start_date.timestamp())
end_ts = int(end_date.timestamp())

# Generate random dates between 2000 and 2024
df['random_date'] = pd.Series(
    pd.to_datetime(
        np.random.randint(start_ts, end_ts, size=len(df)),
        unit='s'
    )
).dt.strftime('%Y-%m-%d')

print(df)
```

```python
# New columns to assign random engagement statistics (clicks)

df['random_views'] = np.random.randint(5000, 12001, size=len(df))

print(df)

df.to_csv("dummydata.csv", index=False)
```
